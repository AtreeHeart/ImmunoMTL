```python
mhcflurry = pd.read_csv("data/Data_S3.csv")
mhcflurry
```

```python
mhcflurry = mhcflurry[(mhcflurry["allele"].str.contains('HLA')) & (mhcflurry["allele"].str.contains(':')) &(mhcflurry["allele"].str.contains('D')==False) & (mhcflurry["allele"].str.contains('o')==False) & (mhcflurry["allele"].str.contains('E')==False)& (mhcflurry["allele"].str.contains('G')==False)]
mhcflurry = mhcflurry[mhcflurry["measurement_kind"] == "mass_spec"]
#mhcflurry_affinity = mhcflurry[(mhcflurry["measurement_kind"] == "affinity") & (mhcflurry["measurement_value"] <= 100) & (mhcflurry["measurement_inequality"] != ">")]
#mhcflurry = pd.concat([mhcflurry_mass, mhcflurry_affinity], ignore_index=True)
pep_len=[]
for i in mhcflurry["peptide"]:
    pep_len.append(len(i))
mhcflurry["len"] = pep_len
mhcflurry = mhcflurry[(mhcflurry["len"]>7) & (mhcflurry["len"]<12)]
mhcflurry
```

```python
tmp_df = mhcflurry.groupby(['allele', 'len']).count()
tmp_df["peptide"]

filtered_df = tmp_df[tmp_df > 10]
filtered_df2 = filtered_df[filtered_df.peptide.notnull()]

final_hla = filtered_df2.index.get_level_values('allele').unique()
final_hla_lst = []
for i in final_hla:
    if filtered_df2.index.get_level_values(0).tolist().count(i) == 4:
        final_hla_lst.append(i)
print("Total HLA allele types: " + str(len(final_hla_lst)))
mhc_df_filtered =  mhcflurry[mhcflurry["allele"].isin(final_hla_lst)]

#filtered_df2
mhc_df_filtered
mhc_df_filtered.to_csv(("../HLA/MHCflurry_filtered.csv"))
```

```python
def entropy(df):
    res = []
    for i in range(len(df.columns)-2):
        AA_unique = df[i+1].unique()
        entropy_ls = []
        for y in range(len(AA_unique)):
            p = df[i+1].tolist().count(AA_unique[y])/len(df[i+1])
            entropy_ls.append( p * math.log(p,2))
        #entropy_sum = -1 * sum(entropy_ls)
        entropy_sum = math.log(20, 2)- (-1*sum(entropy_ls) + (1/np.log(2)*(20-1)/(2*len(df[i+1]))))
        
        res.append(entropy_sum)    
    return res
```

```python
def normalized(lst):
    norm_lst = []
    for i in lst:
        norm_lst.append(1 - i/sum(lst))
    return norm_lst
```

```python
np_8mer = np.empty(shape=8)
np_9mer = np.empty(shape=9)
np_10mer = np.empty(shape=10)
np_11mer = np.empty(shape=11)

hla_unique = []
for i in range(8,12):
    df = mhc_df_filtered[mhc_df_filtered["len"]==i]
    for y in mhc_df_filtered["allele"].unique():
        hla_groupby = df.groupby(["allele"])
        peplist = hla_groupby.get_group(y)
        split_pep = peplist["peptide"].str.split("", expand = True)
        entropy_res = normalized(entropy(split_pep))
        #print(y + "_" + str(i))
        #print(entropy_res)
        #print(len(peplist))
        if i == 8:
            np_8mer = np.vstack((np_8mer,entropy_res))
            hla_unique.append(y)
        elif i == 9:
            np_9mer = np.vstack((np_9mer,entropy_res))
        elif i == 10:
            np_10mer = np.vstack((np_10mer,entropy_res))
        elif i == 11:
            np_11mer = np.vstack((np_11mer,entropy_res))
np_8mer = np.delete(np_8mer, (0), axis = 0)
np_9mer = np.delete(np_9mer, (0), axis = 0)
np_10mer = np.delete(np_10mer, (0), axis = 0)
np_11mer = np.delete(np_11mer, (0), axis = 0)
np_kmer = np.concatenate((np_8mer, np_9mer, np_10mer, np_11mer), axis=1)
training = pd.DataFrame(np_kmer)
training.index = hla_unique
#training.to_csv("../HLA/MHCflurry_training_reverse.csv")
training
```

```python
K = range(2,10)
plt.figure(figsize=(4, 3), dpi=300)
plt.rcParams['font.weight'] = 600


silhouette_avg = []
for k in K:
    kmeans_fit = KMeans(n_clusters = k).fit(training)
    silhouette_avg.append(silhouette_score(training, kmeans_fit.labels_))
plt.plot(K, silhouette_avg, 'bx-')
plt.xlabel('Clusters',fontweight=600)
plt.title('Silhouette Coefficient', fontweight=600)
plt.show()

sil_lst = []
for i in range(len(silhouette_avg)):
    if i + 1 >= len(silhouette_avg):
        break
    sil_lst.append(silhouette_avg[i]-silhouette_avg[i+1])
print({sil_lst.index(max(sil_lst))})

Sum_of_squared_distances = []
for k in K:
    km = KMeans(n_clusters=k)
    km = km.fit(training)
    Sum_of_squared_distances.append(km.inertia_)

plt.figure(figsize=(4, 3), dpi=300)
plt.rcParams['font.weight'] = 600

plt.plot(K, Sum_of_squared_distances, 'bx-')
plt.xlabel('k',fontweight=600)
plt.ylabel('Sum_of_squared_distances',fontweight=600)
plt.title('Elbow Method For Optimal k',fontweight=600)
plt.show()
```

```python
def kmeans_pred(i, df):
    kmeans = KMeans(n_clusters=i, random_state = 30).fit(df)
    model_filename = 'kmeans_HLAcluster{}.pkl'.format(i)
    with open(model_filename, 'wb') as file:
        pickle.dump(kmeans, file)

    # Compute cluster centers and predict cluster indices
    clusters = kmeans.predict(df)
    cluster_res = pd.DataFrame(df.index, clusters)
    #print(kmeans.cluster_centers_)
    unique_elements, counts_elements = np.unique(clusters, return_counts=True)
    #print(clusters)
    print("Count of each clusters:")
    print(np.asarray((unique_elements, counts_elements)))
    
    """Find centroid"""
    closest_pt_idx = []
    tr_array = training.to_numpy()
    for iclust in range(kmeans.n_clusters):
        cluster_pts = tr_array[kmeans.labels_ == iclust]
        cluster_pts_indices = np.where(kmeans.labels_ == iclust)[0]
        #print(kmeans.cluster_centers_[iclust])
        #print(cluster_pts_indices)
        min_idx = np.argmin([euclidean(tr_array[[idx],:], kmeans.cluster_centers_[iclust]) for idx in cluster_pts_indices])
        #print(cluster_pts_indices)
        #print(min_idx)
        print('closest index to cluster ' + str(iclust) + ' center: ', training.index[cluster_pts_indices[min_idx]])
        
    # Define our own color map
    LABEL_COLOR_MAP = {0 : '#549CAD',1 : '#B3C05E',2 : '#EC5B4B', 3:'#CC9418', 4:'#7F7F7F', 5:'y',6:'#475246',7:'gray'}
    label_color = [LABEL_COLOR_MAP[l] for l in clusters]

    # Plot the scatter digram
    plt.figure(figsize = (7,7))
    # Set x-axis label with a larger font size
    plt.xlabel('PC 1', fontsize=18, fontweight='bold')
    # Set y-axis label with a larger font size
    plt.ylabel('PC 2', fontsize=18, fontweight='bold')

    # Increase font size of the tick labels on both axes
    plt.tick_params(axis='both', which='major', labelsize=16)
    plt.rcParams["axes.linewidth"] = 2
    plt.xlim(-0.6, 0.6)
    plt.ylim(-0.4, 0.6)
    plt.scatter(pca_df[:,0],pca_df[:,1], c= label_color, alpha=1, edgecolors='#747474', s=80) 
    #print(*scatter.legend_elements())
    #plt.add_artist(legend1)

    plt.show()
    #filename = "hla/figures/" + str(i) + "clusters.png"
    #plt.savefig(filename)
    
    return clusters
```

```python
AA_list = ['A', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'K', 'L', 'M', 'N', 'P', 'Q', 'R', 'S', 'T', 'V', 'W', 'Y']

def seqlogo(df, HLA):
    cent_df = df[df["allele"] == HLA]
    for i in range(8,12):
        len_group = cent_df.groupby('len')
        peplist = len_group.get_group(i)["peptide"]
        split_pep = peplist.str.split("", expand = True)
        
        split_pep = split_pep.drop([0,i+1], axis=1)
        split_pep.columns = range(0,i)
        #print(split_pep)
        pos_AA_p = pd.DataFrame(columns = AA_list)
        for y in range(0,i):
            AA_p = []
            entropy_lst = []
            for aa in range(len(AA_list)):
                p = (split_pep[y].to_list().count(AA_list[aa]))/len(peplist)
                AA_p.append(p)
                if p != 0:
                    entropy_lst.append(p * math.log(p,2))
            R_ = math.log(20, 2)- (-1*sum(entropy_lst) + (1/np.log(2)*(20-1)/(2*len(peplist))))

            pos_AA_p.loc[len(pos_AA_p)] = [item * R_ for item in AA_p]
        
        # create Logo object
        ww_logo = logomaker.Logo(pos_AA_p)
        
        plt.show()
```
