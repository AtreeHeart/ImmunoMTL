```python
import pandas as pd
import numpy as np
import math
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import StandardScaler
from scipy.spatial.distance import euclidean
import matplotlib.pyplot as plt
import seqlogo
import logomaker
import pickle
```

```python
mhcflurry_df = pd.read_csv("../HLA/Data_S3.csv")
# data downloaded from https://data.mendeley.com/datasets/zx3kjzc3yx/3
```

```python
# Filter valid HLA alleles 
mhcflurry_filtered = mhcflurry_df[
    mhcflurry_df["allele"].str.contains("HLA") &
    mhcflurry_df["allele"].str.contains(":") &
    ~mhcflurry_df["allele"].str.contains("D|o|E|G")
]

# Keep only mass spectrometry data and 8–11-mer peptides
mhcflurry_filtered = mhcflurry_filtered.copy()
mhcflurry_filtered["len"] = mhcflurry_filtered["peptide"].str.len()
mhcflurry = mhcflurry_filtered[
    (mhcflurry_filtered["measurement_kind"] == "mass_spec") &
    (mhcflurry_filtered["len"].between(8, 11))
]

# Select HLA types with enough peptides for all 8–11-mer lengths
tmp_df = mhcflurry.groupby(['allele', 'len']).count()
filtered_df = tmp_df[tmp_df["peptide"] > 10]

final_hla = filtered_df.index.get_level_values('allele').unique()
final_hla_lst = [
    hla for hla in final_hla
    if filtered_df.index.get_level_values(0).tolist().count(hla) == 4
]

print(f"Tier 1 HLA candidates (n={len(final_hla_lst)}):\n", final_hla_lst)
mhc_df_filtered = mhcflurry[mhcflurry["allele"].isin(final_hla_lst)]
mhc_df_filtered.to_csv("../HLA/MHCflurry_filtered.csv")
tier1_hlas = set(final_hla_lst)
```

```python
mhcflurry_filtered = mhcflurry_filtered.copy()
mhcflurry_filtered["len"] = mhcflurry_filtered["peptide"].str.len()

# Select MS 9-mer data
mass_spec_9mer = mhcflurry_filtered[
    (mhcflurry_filtered["measurement_kind"] == "mass_spec") & 
    (mhcflurry_filtered["len"] == 9)
]

# Select quantitative affinity binders (IC50 <= 500 nM)
affinity_df = mhcflurry_filtered[mhcflurry_filtered["measurement_kind"] == "affinity"]
strong_binders = affinity_df[
    ((affinity_df["measurement_inequality"] == "=") & (affinity_df["measurement_value"] <= 500)) |
    ((affinity_df["measurement_inequality"] == "<") & (affinity_df["measurement_value"] <= 500))
]
affinity_9mer = strong_binders[strong_binders["peptide"].str.len() == 9]

# Combine MS and affinity data for 9-mers
tier2_pool = pd.concat([mass_spec_9mer, affinity_9mer])

hla_counts = tier2_pool.groupby("allele")["peptide"].count()
tier2_hlas = set(hla_counts[hla_counts > 10].index)

tier2_hlas = tier2_hlas - tier1_hlas
print(f"Tier 2 HLA candidates (n={len(tier2_hlas)}):\n", tier2_hlas)

tier2_mhc_df = mhcflurry_filtered[mhcflurry_filtered["allele"].isin(tier2_hlas)]
tier2_mhc_df
```

```python
def entropy(df):
    res = []
    for i in range(len(df.columns)-2):
        AA_unique = df[i+1].unique()
        entropy_ls = []
        for y in range(len(AA_unique)):
            p = df[i+1].tolist().count(AA_unique[y])/len(df[i+1])
            entropy_ls.append( p * math.log(p,2))
        #entropy_sum = -1 * sum(entropy_ls)
        entropy_sum = math.log(20, 2)- (-1*sum(entropy_ls) + (1/np.log(2)*(20-1)/(2*len(df[i+1]))))
        
        res.append(entropy_sum)    
    return res
```

```python
def normalized(lst):
    norm_lst = []
    for i in lst:
        norm_lst.append(1 - i/sum(lst))
    return norm_lst
```

```python
np_8mer = np.empty(shape=8)
np_9mer = np.empty(shape=9)
np_10mer = np.empty(shape=10)
np_11mer = np.empty(shape=11)

hla_unique = []
for i in range(8,12):
    df = mhc_df_filtered[mhc_df_filtered["len"]==i]
    for y in mhc_df_filtered["allele"].unique():
        hla_groupby = df.groupby(["allele"])
        peplist = hla_groupby.get_group(y)
        split_pep = peplist["peptide"].str.split("", expand = True)
        entropy_res = normalized(entropy(split_pep))
        #print(y + "_" + str(i))
        #print(entropy_res)
        #print(len(peplist))
        if i == 8:
            np_8mer = np.vstack((np_8mer,entropy_res))
            hla_unique.append(y)
        elif i == 9:
            np_9mer = np.vstack((np_9mer,entropy_res))
        elif i == 10:
            np_10mer = np.vstack((np_10mer,entropy_res))
        elif i == 11:
            np_11mer = np.vstack((np_11mer,entropy_res))
np_8mer = np.delete(np_8mer, (0), axis = 0)
np_9mer = np.delete(np_9mer, (0), axis = 0)
np_10mer = np.delete(np_10mer, (0), axis = 0)
np_11mer = np.delete(np_11mer, (0), axis = 0)
np_kmer = np.concatenate((np_8mer, np_9mer, np_10mer, np_11mer), axis=1)

training = pd.DataFrame(np_kmer)
training.index = hla_unique
training.to_csv("../HLA/MHCflurry_training.csv")
training
```

```python
def kmeans_pred(i, df):
    kmeans = KMeans(n_clusters=i, random_state=30).fit(df)
    model_filename = 'kmeans_HLAcluster{}.pkl'.format(i)
    with open(model_filename, 'wb') as file:
        pickle.dump(kmeans, file)

    # Compute cluster centers and predict cluster indices
    clusters = kmeans.predict(df)
    cluster_res = pd.DataFrame(df.index, clusters)
    unique_elements, counts_elements = np.unique(clusters, return_counts=True)
    print("Count of each cluster:")
    print(np.asarray((unique_elements, counts_elements)))

    """Find centroid"""
    closest_pt_idx = []
    tr_array = training.to_numpy()
    for iclust in range(kmeans.n_clusters):
        cluster_pts = tr_array[kmeans.labels_ == iclust]
        cluster_pts_indices = np.where(kmeans.labels_ == iclust)[0]
        min_idx = np.argmin([euclidean(tr_array[[idx],:], kmeans.cluster_centers_[iclust]) for idx in cluster_pts_indices])
        print('closest index to cluster ' + str(iclust) + ' center: ', training.index[cluster_pts_indices[min_idx]])

    # Define color map based on the number of clusters
    color_palette = ['#549CAD', '#B3C05E', '#EC5B4B', '#CC9418', '#7F7F7F', 'purple', '#475246', 'gray']
    LABEL_COLOR_MAP = {k: color_palette[k] for k in range(i)}
    label_color = [LABEL_COLOR_MAP[l] for l in clusters]

    # Plot the scatter diagram
    plt.figure(figsize=(7, 7))
    plt.xlabel('PC 1', fontsize=18, fontweight='bold', labelpad=15)
    plt.ylabel('PC 2', fontsize=18, fontweight='bold', labelpad=20)
    plt.tick_params(axis='both', which='major', labelsize=16)
    plt.rcParams["axes.linewidth"] = 2
    plt.xlim(-0.6, 0.6)
    plt.ylim(-0.4, 0.6)
    scatter = plt.scatter(pca_df[:, 0], pca_df[:, 1], c=label_color, alpha=1, edgecolors='#747474', s=80)

    # Add legend
    legend_labels = {v: k + 1 for k, v in LABEL_COLOR_MAP.items()}  # Adjust cluster numbers to start from 1
    handles = [plt.Line2D([0], [0], marker='o', color='w', markerfacecolor=color, markersize=10, label=f'Cluster {legend_labels[color]}') for color in LABEL_COLOR_MAP.values()]
    legend = plt.legend(handles=handles, title='Clusters', fontsize=14, title_fontsize=16, loc='upper left', bbox_to_anchor=(1, 1), frameon=False)
    plt.title('K-Means clustering of HLA types (K={})'.format(i), fontsize=20, fontweight='bold', y=1.1)

    # Save the plot with 300 dpi and transparent background
    filename = "kmeans_clusters_k{}.png".format(i)
    #plt.savefig(filename, dpi=300, transparent=True, bbox_inches='tight')
    filename = "kmeans_clusters_k{}.tiff".format(i)
    #plt.savefig(filename, dpi=300, format='tiff', transparent=True, bbox_inches='tight')

    plt.show()
    
    return clusters
```

```python
from scipy.spatial.distance import cosine
from scipy.stats import pearsonr, spearmanr

def assign_by_similarity(tier1_df, tier2_df, tier1_cluster_df, method="cosine"):
    results = []
    for t2_hla in tier2_df.index:
        t2_vec = tier2_df.loc[t2_hla].values

        best_score = -np.inf
        best_hla = None

        for t1_hla in tier1_df.index:
            t1_vec = tier1_df.loc[t1_hla].values

            if method == "cosine":
                score = 1 - cosine(t1_vec, t2_vec)  # cosine similarity
            elif method == "pearson":
                score, _ = pearsonr(t1_vec, t2_vec)
            elif method == "spearman":
                score, _ = spearmanr(t1_vec, t2_vec)
            else:
                raise ValueError("Unsupported method: choose cosine, pearson, or spearman")

            if score > best_score:
                best_score = score
                best_hla = t1_hla
        if "HLA" in tier1_cluster_df.columns:
            tier1_cluster_df = tier1_cluster_df.set_index("HLA")
        assigned_cluster = tier1_cluster_df.loc[best_hla, "cluster"]
        results.append({"HLA": t2_hla, "assigned_cluster": assigned_cluster, "best_match": best_hla, "similarity": best_score})

    return pd.DataFrame(results)
```

```python
def kmeans_pred2(df, hla_clusters_df, pca_df_pd):
    
    # Merge the clusters with HLA clusters dataframe
    pca_clusters_df = pd.DataFrame({'HLA': df.index})
    merged_df = pd.merge(pca_clusters_df, hla_clusters_df, on='HLA', how='left')
    #print(merged_df)
    # Define color map based on the HLA clusters
    #color_palette = ['#549CAD', '#B3C05E', '#EC5B4B', '#CC9418', '#7F7F7F', 'purple', 'blue']
    color_palette = ['#464646', '#549CAD', '#B3C05E', '#EC5B4B', '#CC9418', '#7F7F7F', 'purple', '#475246']

    LABEL_COLOR_MAP = {k: color_palette[k] for k in range(len(color_palette))}
    LABEL_COLOR_MAP['not-defined'] = 'black'  # Color for not defined clusters
    
    # Assign colors, handling NaN values
    merged_df['color'] = merged_df['cluster'].apply(lambda x: LABEL_COLOR_MAP.get(x, 'black'))
    #print(merged_df[['HLA', 'cluster', 'color']])  # Debug print to check color assignment

    # Plot the scatter diagram
    # Plot the scatter diagram
    plt.figure(figsize=(7, 7))
    plt.xlabel('PC 1', fontsize=18, fontweight='bold')
    plt.ylabel('PC 2', fontsize=18, fontweight='bold')
    plt.tick_params(axis='both', which='major', labelsize=16)
    plt.rcParams["axes.linewidth"] = 2
    plt.xlim(-0.6, 0.6)
    plt.ylim(-0.4, 0.6)
    scatter = plt.scatter(pca_df_pd.iloc[:, 0], pca_df_pd.iloc[:, 1], c=merged_df['color'], alpha=1, edgecolors='#747474', s=80)

    # Add legend
    unique_colors = merged_df['color'].unique()
    legend_labels = {color: f'Group {k}' if k != 'not-defined' else 'Not-defined' for k, color in LABEL_COLOR_MAP.items() if color in unique_colors}
    handles = [plt.Line2D([0], [0], marker='o', color='w', markerfacecolor=color, markersize=10, label=label) for color, label in legend_labels.items()]
    legend = plt.legend(handles=handles, title='Prime \nGroups', fontsize=14, title_fontsize=16, loc='upper left', bbox_to_anchor=(1, 1), frameon=False)
    plt.title('Group of HLA-I alleles defined by PRIME', fontsize=20, fontweight='bold', y=1.1)

    # Save the plot with 300 dpi and transparent background
    filename = "HLA_groups_prime.png"
    plt.savefig(filename, dpi=300, transparent=True, bbox_inches='tight')
    filename = "HLA_groups_prime.tiff"
    plt.savefig(filename, dpi=300, format='tiff', transparent=True, bbox_inches='tight')
    plt.show()
    
    pass
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.metrics import silhouette_score

# Define the range of epsilon values to explore
eps_values = np.linspace(0.1, 2, 20)  # Adjust the range and step size as needed

plt.figure(figsize=(4, 3), dpi=300)
plt.rcParams['font.weight'] = 600

silhouette_avg = []

# Loop over different values of eps
for eps in eps_values:
    dbscan_fit = DBSCAN(eps=eps, min_samples=5).fit(training)  # min_samples can be adjusted
    labels = dbscan_fit.labels_
    # Only calculate silhouette score if more than 1 cluster is found
    if len(set(labels)) > 1:
        silhouette_avg.append(silhouette_score(training, labels))
    else:
        silhouette_avg.append(-1)  # Use -1 to indicate a poor clustering

plt.plot(eps_values, silhouette_avg, 'bx-')
plt.xlabel('Epsilon (eps)', fontweight=600)
plt.title('Silhouette Coefficient', fontweight=600)
filename = "silhouette_dbscan.png"
#plt.savefig(filename, dpi=300, transparent=True, bbox_inches='tight')
plt.show()

sil_lst = []
for i in range(len(silhouette_avg)):
    if i + 1 >= len(silhouette_avg):
        break
    sil_lst.append(silhouette_avg[i] - silhouette_avg[i+1])
optimal_eps = eps_values[sil_lst.index(max(sil_lst))]
```

```python
import pickle
import numpy as np
import pandas as pd
from sklearn.cluster import DBSCAN
from sklearn.metrics import silhouette_score
from scipy.spatial.distance import euclidean
import matplotlib.pyplot as plt

def dbscan_pred(eps, df):
    dbscan = DBSCAN(eps=eps, min_samples=5).fit(df)
    model_filename = 'dbscan_HLAcluster_eps{}.pkl'.format(eps)
    with open(model_filename, 'wb') as file:
        pickle.dump(dbscan, file)

    # Compute cluster labels
    clusters = dbscan.labels_
    unique_elements, counts_elements = np.unique(clusters, return_counts=True)
    print("Count of each cluster (including noise):")
    print(np.asarray((unique_elements, counts_elements)))

    """Find centroid"""
    closest_pt_idx = []
    tr_array = df.to_numpy()
    for iclust in np.unique(clusters):
        if iclust == -1:
            continue  # Skip noise points
        cluster_pts = tr_array[clusters == iclust]
        cluster_pts_indices = np.where(clusters == iclust)[0]
        min_idx = np.argmin([euclidean(tr_array[[idx],:], dbscan.components_[iclust]) for idx in cluster_pts_indices])
        print('Closest index to cluster ' + str(iclust) + ' center: ', df.index[cluster_pts_indices[min_idx]])

    # Define color map based on the number of clusters
    unique_clusters = np.unique(clusters)
    color_palette = ['#549CAD', '#B3C05E', '#EC5B4B', '#CC9418', '#7F7F7F', 'purple', '#475246', 'gray']
    LABEL_COLOR_MAP = {k: color_palette[k % len(color_palette)] for k in unique_clusters if k != -1}
    LABEL_COLOR_MAP[-1] = 'black'  # Noise points in black
    label_color = [LABEL_COLOR_MAP[l] for l in clusters]

    # Plot the scatter diagram
    plt.figure(figsize=(7, 7))
    plt.xlabel('PC 1', fontsize=18, fontweight='bold')
    plt.ylabel('PC 2', fontsize=18, fontweight='bold')
    plt.tick_params(axis='both', which='major', labelsize=16)
    plt.rcParams["axes.linewidth"] = 2
    plt.xlim(-0.6, 0.6)
    plt.ylim(-0.4, 0.6)
    scatter = plt.scatter(pca_df[:, 0], pca_df[:, 1], c=label_color, alpha=1, edgecolors='#747474', s=80)

    # Add legend
    legend_labels = {v: k + 1 for k, v in LABEL_COLOR_MAP.items()}  # Adjust cluster numbers to start from 1
    handles = [plt.Line2D([0], [0], marker='o', color='w', markerfacecolor=color, markersize=10, label=f'Cluster {legend_labels[color]}') for color in LABEL_COLOR_MAP.values()]
    legend = plt.legend(handles=handles, title='Clusters', fontsize=14, title_fontsize=16, loc='upper left', bbox_to_anchor=(1, 1), frameon=False)
    plt.title('DBSCAN clustering of HLA types (eps={})'.format(eps), fontsize=20, fontweight='bold', y=1.1)

    # Save the plot with 300 dpi and transparent background
    filename = "dbscan_clusters_eps{}.png".format(eps)
    plt.savefig(filename, dpi=300, transparent=True, bbox_inches='tight')
    filename = "dbscan_clusters_eps{}.tiff".format(eps)
    plt.savefig(filename, dpi=300, format='tiff', transparent=True, bbox_inches='tight')

    plt.show()

    return clusters
```

```python
import pickle
import numpy as np
import pandas as pd
from sklearn.mixture import GaussianMixture
from sklearn.metrics import silhouette_score
from scipy.spatial.distance import euclidean
import matplotlib.pyplot as plt

def gmm_pred(df):
    # Define the range of components to explore
    components_range = range(3, 10)
    silhouette_avg = []

    # Loop over different numbers of components
    for n_components in components_range:
        gmm = GaussianMixture(n_components=n_components, random_state=30).fit(df)
        labels = gmm.predict(df)
        silhouette_avg.append(silhouette_score(df, labels))

    # Find the optimal number of components
    optimal_n_components = components_range[silhouette_avg.index(max(silhouette_avg))]

    # Perform GMM clustering with the optimal number of components
    gmm = GaussianMixture(n_components=optimal_n_components, random_state=30).fit(df)
    model_filename = 'gmm_HLAcluster_n{}.pkl'.format(optimal_n_components)
    with open(model_filename, 'wb') as file:
        pickle.dump(gmm, file)

    # Compute cluster labels
    clusters = gmm.predict(df)
    unique_elements, counts_elements = np.unique(clusters, return_counts=True)
    print("Count of each cluster:")
    print(np.asarray((unique_elements, counts_elements)))

    """Find centroid"""
    closest_pt_idx = []
    tr_array = df.to_numpy()
    for iclust in range(gmm.n_components):
        cluster_pts = tr_array[clusters == iclust]
        cluster_pts_indices = np.where(clusters == iclust)[0]
        min_idx = np.argmin([euclidean(tr_array[[idx],:], gmm.means_[iclust]) for idx in cluster_pts_indices])
        print('Closest index to cluster ' + str(iclust) + ' center: ', df.index[cluster_pts_indices[min_idx]])

    # Define color map based on the number of clusters
    unique_clusters = np.unique(clusters)
    color_palette = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b', '#e377c2', '#7f7f7f']
    LABEL_COLOR_MAP = {k: color_palette[k % len(color_palette)] for k in unique_clusters}
    label_color = [LABEL_COLOR_MAP[l] for l in clusters]

    # Plot the scatter diagram
    plt.figure(figsize=(7, 7))
    plt.xlabel('PC 1', fontsize=18, fontweight='bold')
    plt.ylabel('PC 2', fontsize=18, fontweight='bold')
    plt.tick_params(axis='both', which='major', labelsize=16)
    plt.rcParams["axes.linewidth"] = 2
    plt.xlim(-0.6, 0.6)
    plt.ylim(-0.4, 0.6)
    scatter = plt.scatter(pca_df[:, 0], pca_df[:, 1], c=label_color, alpha=1, edgecolors='#747474', s=80)

    # Add legend
    legend_labels = {v: k + 1 for k, v in LABEL_COLOR_MAP.items()}  # Adjust cluster numbers to start from 1
    handles = [plt.Line2D([0], [0], marker='o', color='w', markerfacecolor=color, markersize=10, label=f'Cluster {legend_labels[color]}') for color in LABEL_COLOR_MAP.values()]
    legend = plt.legend(handles=handles, title='Clusters', fontsize=14, title_fontsize=16, loc='upper left', bbox_to_anchor=(1, 1), frameon=False)
    plt.title('GMM clustering of HLA types (n={})'.format(optimal_n_components), fontsize=20, fontweight='bold', y=1.1)

    # Save the plot with 300 dpi and transparent background
    filename = "gmm_clusters_n{}.png".format(optimal_n_components)
    plt.savefig(filename, dpi=300, transparent=True, bbox_inches='tight')
    filename = "gmm_clusters_n{}.tiff".format(optimal_n_components)
    #plt.savefig(filename, dpi=300, format='tiff', transparent=True, bbox_inches='tight')

    plt.show()

    return clusters
```

```python
final_pred_gmm = gmm_pred(training)
final_pred_gmmdf = pd.DataFrame({"HLA":training.index,"cluster": final_pred_gmm})
final_pred_gmmdf.to_csv("../HLA/gmm_clusters_n4.csv")
```
